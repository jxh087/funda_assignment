<a id='Data'></a>
## 0. Data Preparation

### 0.1 Load the Data

```python
# Load the data
df = pd.read_csv("Property Sales of Melbourne City.csv")
print("The original dataset has {0} observations".format(df.shape[0]))
df.head(5)
```

### 0.2 Check for missing values

```python
# Check for missing values
missing_values_percentage = (df.isnull().sum() / len(df)) * 100
missing_values_percentage = missing_values_percentage.sort_values(ascending=False)

print("\nMissing Values:")
for column, percentage in missing_values_percentage.items():
    print(f"{column}: {percentage:.2f}%")
```

### 0.3 Check Duplicate values

```python
# Check for duplicate entries
print("\nDuplicate Entries:")
print(df.duplicated().sum())
```

### 0.4 Select useful columns

```python
# Drop the columns base on the percentage of missing values and redundant information.
columns_to_drop = [
    "Unnamed: 0",
    "Address",
    "Rooms",
    "Bedroom2",
    "Bathroom",
    "Car",
    "Landsize",
    "BuildingArea",
    "YearBuilt",
    "CouncilArea",
    "Lattitude",
    "Longtitude",
]
df = df.drop(columns=columns_to_drop)
```

### 0.5 Describe the selected data

```python
df.describe().round(2)
```

### 0.6 Extract the Date

```python
# Extracting year, month, and generating year-quarter
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Quarter"] = df["Date"].dt.quarter
df["Year_Quarter"] = df["Year"].astype(str) + "-Q" + df["Quarter"].astype(str)

df.to_csv("Property_Sales.csv")
```

<a id='Market Share'></a>
## 1.0 Market Share

```python
# Aggregate the df to find the top 10 agents
# We will consider both the number of sales and the total sales value
top_agents = df.groupby('SellerG').agg({'Price': ['count', 'sum']}).sort_values(by=('Price', 'sum'), ascending=False)
top_agents.columns = ['Number_of_Sales', 'Total_Sales_Value']
top_10_agents = top_agents.head(10).index.tolist()

# Mark the top 10 agents in the original dfset and label others as 'Other'
df['Top_Agent'] = df['SellerG'].apply(lambda x: x if x in top_10_agents else 'Other')
```

```python
# Use DuckDB to generare a table
csv_file_path = "Property_Sales.csv"

conn = duckdb.connect(database=":memory:", read_only=False)

# Load the CSV file into a DuckDB table
conn.execute(
    "CREATE TABLE property_sales AS SELECT * FROM read_csv_auto('{}')".format(
        csv_file_path
    )
)

# Run a SQL query on the loaded data

query = "SELECT SellerG, count(*) AS Total_Number,round((count(*) / (SELECT count(*) FROM property_sales)) * 100,1) AS Percentage_of_number,SUM(Price) AS Total_Sales_AUD, round((SUM(Price) / (SELECT SUM(Price) FROM Property_sales)) * 100,1) AS Percentage_of_Value FROM Property_sales GROUP BY SellerG order by Percentage_of_Value desc LIMIT 10"

conn.sql(query).show()

result_df = conn.sql(query).df()
```

## 2.  Sales Trend

### 2.1  Sales Trend Over Time (AUD)

```python
# Filter the dfset for only the top 10 agencies, excluding "Others"
df_top_10 = df[df['SellerG'].isin(top_10_agents)]

# Group by Year_Quarter and SellerG, then sum the sales
sales_trend_top_10 = df_top_10.groupby(['Year_Quarter', 'SellerG']).agg(Total_Sales=('Price', 'sum')).reset_index()

# Pivot the table for plotting
sales_trend_pivot_top_10 = sales_trend_top_10.pivot(index='Year_Quarter', columns='SellerG', values='Total_Sales')

# Plotting the sales trend over Year_Quarter for only the top 10 agencies
plt.figure(figsize=(14, 8))
for agency in sales_trend_pivot_top_10.columns:
    plt.plot(sales_trend_pivot_top_10.index, sales_trend_pivot_top_10[agency],marker='o', label=agency)

plt.title('Sales Trend of Top 10 Agencies', fontsize=18)
plt.xlabel('Time', fontsize=16)
plt.ylabel('Total Sales (AUD)', fontsize=16)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

**Summary**:Q1 is the least active season for the real estate market. The top 3 agencies in the market are Jellis, Nelson and Marshall.

### 2.2 Sales Trend Over Time (Numbers)

```python
# Line Graph - Sales trend over time for the top agencies

# Filter the dfset for only the top 10 agencies, excluding "Others"
df_top_10 = df[df['SellerG'].isin(top_10_agents)]

# Group by Year_Quarter and SellerG, then sum the sales
sales_trend_top_10 = df_top_10.groupby(['Year_Quarter', 'SellerG']).agg(Total_Sales=('Price', 'count')).reset_index()

# Pivot the table for plotting
sales_trend_pivot_top_10 = sales_trend_top_10.pivot(index='Year_Quarter', columns='SellerG', values='Total_Sales')

# Plotting the sales trend over Year_Quarter for only the top 10 agencies
plt.figure(figsize=(14, 8))
for agency in sales_trend_pivot_top_10.columns:
    plt.plot(sales_trend_pivot_top_10.index, sales_trend_pivot_top_10[agency], marker='o', label=agency)

plt.title('Sales Trend of Top 10 Agencies', fontsize=18)
plt.xlabel('Time', fontsize=16)
plt.ylabel('Total Numbers', fontsize=16)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

## 3. Sales Methods and Property Type

```python
# Pie chart for Method distribution
method_distribution = df.groupby("Method").size().reset_index(name="Count")

# Custom function to format the autopct string
def autopct_format(counts):
    def my_format(pct):
        total = sum(counts)
        val = int(round(pct * total / 100.0))
        return '{v:d}\n({p:.1f}%)'.format(p=pct, v=val)
    return my_format

# Pie Chart: Sales Method Distribution for Top Agencies
plt.figure(figsize=(10, 10))
plt.pie(
    method_distribution["Count"],
    labels=method_distribution["Method"],
    autopct=autopct_format(method_distribution["Count"]),
    startangle=140,
)
plt.title("Sales Method Distribution", fontsize=18)
plt.show()
```

**Summary**: - The majority of sales are completed with the 'S-Sold' method (65.4%), indicating successful sales without prior offers or passing in at auctions. The 'SP-Sold Prior' and 'PI-Passed In' also represent significant portions, suggesting strategies like accepting offers before auctions or offer private negotiations are also common practices.

### 3.1 Selling Method (Top 10)

```python
# Filtering for only the top 10 agencies
df_top_10_prices = df[df['SellerG'].isin(top_10_agents)]

# Calculating the percentage of each selling method sold by the top 10 agencies
df_top_10_types = df_top_10_prices.groupby(['SellerG', 'Method']).size().unstack(fill_value=0)
df_top_10_types_percent = df_top_10_types.div(df_top_10_types.sum(axis=1), axis=0) * 100

#Plotting with the y-axis representing the total number of properties sold
ax = df_top_10_types.plot(kind='bar', stacked=True, figsize=(14, 8), colormap='viridis')
plt.title('Method of Properties Sold by Top 10 Agencies', fontsize=18)
plt.xlabel('Agency', fontsize=16)
plt.ylabel('Total Number of Properties Sold', fontsize=16)
plt.xticks(rotation=45)
plt.legend(title='Method')

# Annotation of percentages within each bar
for i, (index, row) in enumerate(df_top_10_types.iterrows()):
    total = row.sum()
    cum_val = 0  # To keep track of the cumulative height of the bars
    for col in df_top_10_types.columns:
        val = row[col]
        if val:
            percentage = (val / total) * 100
            ax.text(i, cum_val + val/2, f'{percentage:.1f}%', ha='center', va='center', color='White', fontsize=8)
            cum_val += val

plt.tight_layout()
plt.show()
```

**Summary**:Specific sale method were favored by individual agencies (e.g., __Marshall__ relys more on PI-Passed In, __Jellis__ relys more on VB-Vendor Bid, __Ray__ and __Barry__ relys more on direct sales, etc.).

### 4.1 Price Distribution (Top 10)

```python
# Correct approach to create a colorful horizontal box plot for the top 10 agencies

# Filtering for only the top 10 agencies
df_top_10_prices = df[df["SellerG"].isin(top_10_agents)]

# Generating colors
colors = plt.cm.viridis(np.linspace(0, 1, len(top_10_agents)))

plt.figure(figsize=(14, 10))

# Creating box plots one by one to apply colors
for i, agency in enumerate(top_10_agents):
    subset = df_top_10_prices[df_top_10_prices["SellerG"] == agency]["Price"]
    plt.boxplot(
        subset,
        positions=[i],
        vert=False,
        patch_artist=True,
        boxprops=dict(facecolor=colors[i], color=colors[i]),
        medianprops=dict(color="yellow"),
        whiskerprops=dict(color=colors[i]),
        capprops=dict(color=colors[i]),
        flierprops=dict(markeredgecolor=colors[i]),
    )

plt.yticks(range(len(top_10_agents)), top_10_agents)
plt.title("Price Distribution for Top 10 Agencies", fontsize=18)
plt.xlabel("Price (AUD)", fontsize=16)
plt.ylabel("Agency", fontsize=16)


# Extracting statistical information for the price distribution of the top 10 agencies
statistics = df_top_10_prices.groupby("SellerG")["Price"].describe().round()

plt.show()
statistics
```

<a id='4.2'></a>
### 4.2 Price vs. Distance

```python
# Log-transforming the 'Price' column
df['Log_Price'] = np.log(df['Price'])

# Preparing df for the bubble chart
bubble_df = df.groupby(['Distance', 'Suburb']) \
                  .agg({'Log_Price': 'mean', 'Propertycount': 'mean'}) \
                  .reset_index()

# Fitting a linear regression model
X = bubble_df[['Distance']]
y = bubble_df['Log_Price']
reg_model = LinearRegression().fit(X, y)
bubble_df['Predicted_Log_Price'] = reg_model.predict(X)

# Linear Regression
linear_regressor = LinearRegression()
linear_regressor.fit(X, y)
y_pred_linear = linear_regressor.predict(X)


# Plotting
plt.figure(figsize=(12, 8))
plt.scatter(
    X, y, s=bubble_df["Propertycount"] / 100, alpha=0.5, label="Original df"
)
plt.plot(X, y_pred_linear, color="red", label="Linear Fit")

plt.title("Sales Price vs. Distance from CBD", fontsize=16)
plt.xlabel("Distance from CBD (km)", fontsize=16)
plt.ylabel("log(Sales Price)", fontsize=16)
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)

plt.tight_layout()
plt.show()

# Outputting the coefficients for explanation
linear_coef_clean = linear_regressor.coef_[0]

linear_coef_clean.round(3)
```

**Summary**:

- Price vs. Distance Trend follows a common trend that property prices decrease as you move away from the city center. The slope of this line is -0.015, which is not really strong corrlation.
- The size of the bubbles indicates how many properties are in each suburb. Larger bubbles in areas closer to the CBD could suggest higher density or popularity,but the price sold are more in the lower side.

## 5 Heatmap

```python
# Heatmap 1: Heatmap of Property Sales by Top 10 Agencies Across Regions
# Filtering only the top 10 agencies
df_top_10 = df[df['SellerG'].isin(top_10_agents)]

# Creating a pivot table for heatmap
pivot_table = df_top_10.pivot_table(values='Price', index='SellerG', columns='Regionname', aggfunc='count', fill_value=0)


# Calculating the average distance for each region
average_distance_per_region = df.groupby('Regionname')['Distance'].mean().sort_values()

# Reordering the columns in the pivot table based on the average distance of regions
ordered_regions = average_distance_per_region.index
pivot_table_ordered = pivot_table.reindex(columns=ordered_regions)

# Plotting the heatmap with ordered regions
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table_ordered, annot=True, fmt="d", cmap='PuBu')
plt.title('Heatmap of Property Sales by Top 10 Agencies Across Regions', fontsize=18)
plt.xlabel('Region Name')
plt.ylabel('Agency', fontsize=16)
plt.xticks(rotation=60)
plt.yticks(rotation=0)

plt.show()

# Heatmap 2: Heatmap of Property Sales by Top 10 Agencies Across Regions
# Creating a pivot table for heatmap
pivot_table = df_top_10.pivot_table(values='Price', index='SellerG', columns='Suburb', aggfunc='count', fill_value=0)


# Calculating the average distance for each region
average_distance_per_region = df.groupby('Suburb')['Distance'].mean().sort_values()

# Reordering the columns in the pivot table based on the average distance of regions
ordered_regions = average_distance_per_region.index
pivot_table_ordered = pivot_table.reindex(columns=ordered_regions)

# Plotting the heatmap with ordered regions
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table_ordered, annot=False, fmt="f", cmap='GnBu')
plt.title('Heatmap of Property Sales by Top 10 Agencies Across Suburbs', fontsize=18)
plt.xlabel('Suburbs')
plt.ylabel('Agency', fontsize=16)
plt.xticks(rotation=90)
plt.yticks(rotation=0)

plt.show()
```

**Summary**: Different agencies focus on Different region/suburbs.

## 6.0 Performance Analysis and Conclusions

```python
# For the final comparison, top 5 agent has been selected.
top_5_agencies = df.groupby('SellerG')['Price'].sum().sort_values(ascending=False).head(5).index

# Filtering df for top 5 agencies
df_top_5 = df[df['SellerG'].isin(top_5_agencies)]

# Calculating metrics
metrics_df = df_top_5.groupby('SellerG').agg(
    Total_Sales=('Price', 'sum'),
    Average_Price=('Price', 'mean'),
    Average_Distance=('Distance', 'mean')
)

# Normalizing Total Sales for comparison
metrics_df['Total_Sales'] = metrics_df['Total_Sales'] / metrics_df['Total_Sales'].max() * 100

# Calculating Market Share
total_sales_top_5 = metrics_df['Total_Sales'].sum()
metrics_df['Market_Share'] = (metrics_df['Total_Sales'] / total_sales_top_5) * 100

# Normalizing other metrics for comparison
metrics_df['Average_Price'] = (metrics_df['Average_Price'] / metrics_df['Average_Price'].max()) * 100
metrics_df['Average_Distance'] = (metrics_df['Average_Distance'] / metrics_df['Average_Distance'].max()) * 100

# Resetting index
metrics_df = metrics_df.reset_index()

# Plotting radar chart
labels=np.array(['Total Sales', 'Market Share', 'Average Price', 'Average Distance'])
num_vars = len(labels)

angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

for index, row in metrics_df.iterrows():
    df = row[['Total_Sales', 'Market_Share', 'Average_Price', 'Average_Distance']].tolist()
    df += df[:1]
    ax.fill(angles, df, alpha=0.25)
    ax.plot(angles, df, label=row['SellerG'])

ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

ax.set_thetagrids(np.degrees(angles[:-1]), labels)

plt.title('Comparative Analysis of Top 5 Agencies', size=18, y=1.1)
plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))

plt.show()

# Displaying the calculated metrics for insight
metrics_df
```

## 6.1 Conclusions

- Marshall specializes in high-value properties, as evidenced by the highest average price, high house sales, suggesting a focus on the luxury segment.
- Jellis leads in total sales and market share, highlighting its dominance in the market and Vendor bid, likely driven by a mix of high-value sells and strategic marketing.
- Nelson and hockingstuart show strong performance but with different strategies, as seen in their average price and distance metrics. Nelson focus on apartment near the CBD and Hockingstuart focus the great mixture of housing profiles.
- Barry, while having the lowest total sales and market share, sells properties that are, on average, furthest from the CBD, indicating a potential focus on suburban or specific regional markets. However, Barry is upcoming agencies which might change the market landscape in the future.
